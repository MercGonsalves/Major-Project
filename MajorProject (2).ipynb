```python
x=df.iloc[:,1:12]
x
```

```python
y=df.iloc[:,12:]
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.85)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.svm import SVC #SVM TEST
svm1=SVC()
svm1.fit(xtrain,ytrain)
ans=svm1.predict(xtest)
ans
```

```python
print("Accuracy on test set: {:.4f}".format(svm1.score(xtest,ytest)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn1=KNeighborsClassifier()
knn1.fit(xtrain,ytrain)
ans1=knn1.predict(xtest)
ans1
```

```python
print("Accuracy on test set: {:.4f}".format(knn1.score(xtest,ytest)))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
ans2=dt.predict(xtest)
ans2
```

```python
print("Accuracy on test set: {:.4f}".format(dt.score(xtest,ytest)))
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=59500)
lr.fit(xtrain,ytrain)
ans3=lr.predict(xtest)
ans3
```

```python
print("Accuracy on test set: {:.4f}".format(lr.score(xtest,ytest)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(xtrain,ytrain)
ans4=rf.predict(xtest)
ans4
```

```python
print("Accuracy on test set: {:.4f}".format(rf.score(xtest,ytest)))
```
